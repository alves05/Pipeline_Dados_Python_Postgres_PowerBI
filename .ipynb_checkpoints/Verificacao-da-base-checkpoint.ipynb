```python
import pandas as pd
import numpy as np
import psycopg2
```

```python
planilha1 = pd.read_excel('online_retail_II.xlsx', sheet_name='Year 2009-2010')
planilha1
```

```python
planilha2 = pd.read_excel('online_retail_II.xlsx', sheet_name='Year 2010-2011')
planilha2
```

```python
data = pd.concat([planilha1, planilha2])
data
```

```python
invoice = []
for i in data['Invoice']:
    invoice.append('IN' + str(i))

date = []
for i in data['InvoiceDate']:
    date.append(i)
    
order_invoice = pd.DataFrame({'Invoice_ID':invoice, 'Date':date})
order_invoice.shape
```

```python
order_invoice['Date_invoice'] = pd.to_datetime(order_invoice['Date']).dt.date
order_invoice['Hour_Invoice'] = pd.to_datetime(order_invoice['Date']).dt.time
order_invoice['Minute_Invoice'] = pd.to_datetime(order_invoice['Date']).dt.minute
order_invoice['Second_Invoice'] = pd.to_datetime(order_invoice['Date']).dt.second
order_invoice
```

```python
client = []
for i in data['Customer ID']:
    client.append(i)
    
country = []
for i in data['Country']:
    country.append(i)
    
order_invoice['Client'] = client
order_invoice['Country'] = country
```
